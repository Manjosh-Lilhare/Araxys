```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('/content/communities-crime-full.csv')
```

```python
df.info()
df.head()
```

```python
df.describe()
```

```python
print("\nDrop rows with missing values:")
print(df.dropna())
```

```python
df = df.drop_duplicates()
print(df)
```

```python
df.replace('?', np.nan, inplace=True)
```

```python
for col in df.columns:
    if df[col].dtype == 'object':
        try:
            df[col] = df[col].astype(float)
        except ValueError:
            pass
```

```python
columns_to_drop = [ 'community', 'communityname', 'fold']
df_cleaned = df.drop(columns=columns_to_drop, axis=1)
```

```python
for col in df_cleaned.columns:
    if df_cleaned[col].dtype in ['float64', 'int64']:
        df_cleaned[col].fillna(df_cleaned[col].median(), inplace=True)
```

```python
df_cleaned.drop_duplicates(inplace=True)
```

```python
df_cleaned.info(), df_cleaned.head()
```

```python
cleaned_file_path = "communities_crime_cleaned.csv"
df_cleaned.to_csv(cleaned_file_path, index=False)

cleaned_file_path
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set style for plots
sns.set_style("whitegrid")
```

```python
# Figure 1: Distribution of Violent Crimes
plt.figure(figsize=(10, 5))
sns.histplot(df_cleaned["ViolentCrimesPerPop"], bins=30, kde=True, color='red')
plt.title("Distribution of Violent Crimes per Population")
plt.xlabel("Violent Crimes Per Population")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(10,6))
correlation = df_cleaned.corr()
top_corr_features = correlation["ViolentCrimesPerPop"].abs().sort_values(ascending=False)[1:21]
sns.heatmap(df_cleaned[top_corr_features.index].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap of Top Features Affecting Crime Rate")
plt.show()
```

```python
# Figure 3: Race Percentage vs Violent Crime Rate
plt.figure(figsize=(10,6))
for race in ["racepctblack", "racePctWhite", "racePctAsian", "racePctHisp"]:
    sns.scatterplot(x=df_cleaned[race], y=df_cleaned["ViolentCrimesPerPop"], label=race, alpha=0.6)
plt.title("Race Percentage vs. Violent Crime Rate")
plt.xlabel("Race Percentage")
plt.ylabel("Violent Crimes Per Population")
plt.legend()
plt.show()
```

```python

## Figure 5: Police Budget vs. Crime Rate
plt.figure(figsize=(10, 5))
sns.regplot(x=df_cleaned["PolicBudgPerPop"], y=df_cleaned["ViolentCrimesPerPop"], scatter_kws={'alpha':0.6}, color='green')
plt.title("Police Budget Per Capita vs. Violent Crime Rate")
plt.xlabel("Police Budget Per Population")
plt.ylabel("Violent Crimes Per Population")
plt.show()
```

```python
plt.figure(figsize=(15, 5))
for race in ["racepctblack", "racePctWhite", "racePctAsian", "racePctHisp"]:
    sns.scatterplot(x=df_cleaned[race], y=df_cleaned["ViolentCrimesPerPop"], label=race, alpha=0.6)
plt.title("Race Percentage vs. Violent Crime Rate")
plt.xlabel("Race Percentage")
plt.ylabel("Violent Crimes Per Population")
plt.legend()
plt.show()
```

```python
key_features = ["racepctblack", "racePctWhite", "PopDens", "PolicBudgPerPop", "PctUsePubTrans", "ViolentCrimesPerPop"]
sns.pairplot(df_cleaned[key_features], diag_kind="kde", corner=True)
plt.show()
```

```python
# Figure 1: Bar Plot - Average Violent Crime Rate by Race Percentage
plt.figure(figsize=(10, 6))
race_features = ["racepctblack", "racePctWhite", "racePctAsian", "racePctHisp"]
avg_crime_by_race = df_cleaned[race_features + ["ViolentCrimesPerPop"]].mean()

sns.barplot(x=race_features, y=avg_crime_by_race[race_features], palette="viridis")
plt.title("Average Race Percentage vs. Violent Crime Rate")
plt.xlabel("Race Group")
plt.ylabel("Average Percentage in Population")
plt.show()

# Figure 2: Histogram (Histplot) - Distribution of Violent Crimes by Race
plt.figure(figsize=(10, 6))
for race in race_features:
    sns.histplot(df_cleaned[race], kde=True, bins=30, label=race, alpha=0.6)

plt.title("Distribution of Race Percentage")
plt.xlabel("Race Percentage")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```
