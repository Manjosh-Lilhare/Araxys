```python


# 1. Visualize Crime Distribution Over Years
crime_types = ['Murder', 'Kidnapping', 'Crime against women', 'Crime against children',
               'Crime Committed by Juveniles', 'Crime against Senior Citizen',
               'Crime against SC', 'Crime against ST', 'Economic Offences', 'Cyber Crimes']

yearly_crime = df.groupby('Year')[crime_types].sum()

plt.figure(figsize=(10, 6))
for crime in crime_types:
    plt.plot(yearly_crime.index, yearly_crime[crime], label=crime)

plt.xlabel('Year')
plt.ylabel('Number of Crimes')
plt.title('Crime Distribution Over Years')
plt.legend()
plt.grid(True)
plt.show()

```

```python
# 2. Visualize City-Wise Crime Rates
city_crime = df.groupby('City')[crime_types].sum().sum(axis=1)

plt.figure(figsize=(10, 6))
plt.bar(city_crime.index, city_crime.values)
plt.xlabel('City')
plt.ylabel('Total Crime')
plt.title('City-Wise Crime Rates')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# 4. Visualize Crime vs. Population
population = df['Population (in Lakhs) (2011)+']
plt.figure(figsize=(10, 6))
for i, crime in enumerate(crime_types):
    plt.subplot(4, 3, i + 1)
    plt.scatter(population, df[crime])
    plt.xlabel('Population (in Lakhs)')
    plt.ylabel(crime)
    plt.title(f'{crime} vs. Population')
plt.tight_layout()
plt.show()
```
